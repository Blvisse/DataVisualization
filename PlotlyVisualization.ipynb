```
data=pd.read_csv('winequality-red.csv')
```

```
data.head()
```

```
sns.heatmap(data.corr(),cmap='RdYlGn',annot=True)
```

```
fig=px.scatter(data,y='alcohol',x='volatile acidity',color='sulphates')
fig.show()
```

```
fig=go.Figure(data=go.Scatter(y=data['fixed acidity'],mode='markers',marker=dict(color=data['density'],showscale=True)))
fig.show()
```

```
colors=['lightslategray',]*6
colors[2]='crimson'
fig=go.Figure(data=[go.Bar(x=data['quality'],y=data['alcohol'],marker_color=colors)])
fig.show()
```

```
 fig=px.bar(data,y='alcohol',x='quality',color='citric acid',title='Quality of wine based on alcohol and citric acid')
fig.show()
```

```
fig=px.pie(data,names='quality')
fig.show()
```

```
fig=px.pie(data,names='quality',values='volatile acidity')
fig.show()
```

```
fig=px.pie(data,names='sulphates',values='chlorides')
fig.show()
```

```
fig=px.pie(data,names='quality',hover_data=['alcohol'])
fig.update_traces(textposition='inside',textinfo='percent+label')
fig.show()
```

```
fig=px.pie(data,names='fixed acidity',values='citric acid')
fig.show()
```

```
fig=go.Figure(data=[go.Pie(labels=data['quality'],values=data['alcohol'],pull=[0.1,0,0,0,0,0])])
fig.show()
```
